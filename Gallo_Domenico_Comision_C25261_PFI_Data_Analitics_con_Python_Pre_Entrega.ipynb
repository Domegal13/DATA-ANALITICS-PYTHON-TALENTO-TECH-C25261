### Conclusi√≥n 1:

### 1.3. Transformaci√≥n de los Datos

#### Limpieza de Datos

#### Se liminan los duplicados

```python
df_ventas = df_ventas.drop_duplicates(keep='first').reset_index(drop=True)
df_ventas.info()
```

#### Se eliminan los registros vacios

```python
df_ventas.dropna(inplace=True)
df_ventas.reset_index(drop=True, inplace=True)
df_ventas.info()
```

```python
df_ventas.sample(4)
```

#### Se convierten los campos  "cantidad" y  "precio"  a int y float

```python
# 'cantidad' a entero
df_ventas['cantidad'] = pd.to_numeric(df_ventas['cantidad'], errors='coerce').fillna(0).astype(int)
```

```python
# Se eliminan los espacios al principio y al final, el simbolo '$'
df_ventas['precio'] = (
    df_ventas['precio']
    .astype(str)
    .str.replace(',', '.')
    .str.replace('$', '', regex=False)
    .str.strip()
    .pipe(pd.to_numeric, errors='coerce')
)

precio_promedio = df_ventas['precio'].mean().round(2)
df_ventas['precio'].fillna(precio_promedio, inplace=True)

df_ventas.info()
```

## 2.- Ventas Mensuales.

#### 2.1. Se crean los campos mes y a√±o

```python
df_ventas['fecha_venta'] = pd.to_datetime(df_ventas['fecha_venta'], format='%d/%m/%Y')
df_ventas['mes'] = df_ventas['fecha_venta'].dt.month
df_ventas['anio'] = df_ventas['fecha_venta'].dt.year
```

```python
display(df_ventas.sample(5))
```

#### 2.2. se crea el campo total_venta

```python
df_ventas['total_venta'] = df_ventas['precio'] * df_ventas['cantidad']
df_ventas.sample(4)
```

#### 2.3. Se calculan las Ventas Mensuales

---



```python
ventas_mensuales = df_ventas.groupby("mes")["total_venta"].sum().reset_index()
ventas_mensuales.rename(columns={'total_venta': 'ventas_mensuales'}, inplace=True)
ventas_mensuales
```

```python
mapear_meses = {
    1: 'Enero',
    2: 'Febrero',
    3: 'Marzo',
    4: 'Abril',
    5: 'Mayo',
    6: 'Junio',
    7: 'Julio',
    8: 'Agosto',
    9: 'Septiembre',
    10: 'Octubre',
    11: 'Noviembre',
    12: 'Diciembre'
}

ventas_mensuales['mes'] = ventas_mensuales['mes'].map(mapear_meses)
ventas_mensuales
```

```python
ventas_mensuales['ventas_mensuales'].describe()
```

```python
mes_menos_ventas = ventas_mensuales.nsmallest(n=1, columns='ventas_mensuales')
mes_mas_ventas = ventas_mensuales.nlargest(n=1, columns='ventas_mensuales')
display("Mes con Menor Ventas")
display(mes_menos_ventas)
display("-"*50)
display('Mes con Mayor Ventas')
display(mes_mas_ventas)
```

### Ventas totales por categoria

```python
ventas_por_categoria = df_ventas.groupby('categoria')['total_venta'].sum().sort_values(ascending=False)

print("--- Ventas Totales por Categor√≠a ---")
print(ventas_por_categoria)
```

#### Ordenar por ventas, de mayor a menor (descendente)

```python

df_ranking_ventas = ventas_mensuales.sort_values(by='ventas_mensuales', ascending=False).reset_index(drop=True)

print("--- Ranking por Ventas (Mayor a Menor) ---")
print(df_ranking_ventas)
```

### Meses con Mayores Ventas

```python
top_ventas_por_mes = df_ranking_ventas.nlargest(n=3, columns='ventas_mensuales').reset_index(drop=True)
print("--- Top 3 Meses con Mayor Venta ---")
display(top_ventas_por_mes)
```

### Ventas Totales por Producto

```python
# ventas_por_producto = df_ventas.groupby('producto')['total_venta'].sum().sort_values(ascending=False)
ventas_por_producto =(df_ventas.groupby('producto', as_index=False)
                                .agg(unidades_vendidas=('cantidad', 'sum'),
                                     ventas_totales=('total_venta', 'sum'))
                                .sort_values(by='ventas_totales', ascending=False)
                                .reset_index(drop=True)).round(2)

print('-'*100)
print("--- Ventas Totales por Producto ---")
print('-'*100)
print(ventas_por_producto.head())
```

### Productos con Mayores Ventas

```python
n=3
top_ventas_por_producto = ventas_por_producto.nlargest(n, 'ventas_totales').reset_index(drop=True)
print('-'*100)
print("--- Top 3 Productos con Mayor Venta ---")
print('-'*100)
display(top_ventas_por_producto)
```

## Combinar los sets de datos de ventas y marketing


```python
display('Ventas')
display(df_ventas.sample(5))
display('Marketing')
display(df_marketing.sample(5))
```

```python
df_marketing['producto'].value_counts().sample(3).reset_index()
```

### Agrupaci√≥n de Marketing por Producto

```python
marketin_producto = (df_marketing.groupby('producto')
                                 .agg(campa√±as=('producto', 'count'),
                                      inversion_total = ('costo', 'sum'))
.sort_values('inversion_total', ascending=False)
.reset_index()).round(2)

print('-'*100)
print("--- Costo de Marketin por Producto ---")
print('-'*100)
print(marketin_producto.head())
```

### Se Combinan los DF "ventas_por_productos" y "marketing_producto"

```python
productos_vta_mkt = pd.merge(marketin_producto, ventas_por_producto, on='producto', how='inner')
productos_vta_mkt.head()
```

### Indicadores Derivados

```python
productos_vta_mkt['ticket_promedio'] = (productos_vta_mkt['ventas_totales'] / productos_vta_mkt['unidades_vendidas']).round(2)
productos_vta_mkt.head(8)
```

```python
productos_vta_mkt['ROAS'] = (productos_vta_mkt['ventas_totales']/productos_vta_mkt['inversion_total']).round(2)
productos_vta_mkt.head(8)
```

```python
productos_vta_mkt = productos_vta_mkt.sort_values('ROAS', ascending=False).reset_index(drop=True)
# productos_vta_mkt.head()
```

###  TOP 3 Productos con Mayor Rendimiento

```python
n = 3
```

```python
productos_con_mayor_rendimiento = productos_vta_mkt.nlargest(n=n, columns='ROAS').reset_index(drop=True)
productos_con_mayor_rendimiento
```

###  TOP 3 Productos con Mayores Ventas

```python
productos_con_mayores_ventas = productos_vta_mkt.nlargest(n, columns='ventas_totales').reset_index(drop=True)
productos_con_mayores_ventas
```

### Resultado:

- Vemos **una fila por producto**, con:
  - `ventas_total`: suma global de ingresos (VENTAS).
  - `unidades_vendidas`: total de unidades vendidas.
  - `inversion_total`: suma global de costos de marketing.
  - `campa√±as`: cu√°ntas campa√±as hubo para ese producto.
  - `ticket_medio`: promedio de venta por unidad.
  - `ROAS`: retorno sobre gasto publicitario (ventas / inversi√≥n).
